```python
from matplotlib import pyplot as plt
import cv2
```

```python
img = cv2. imread(r"C:\Users\Lenovo\Documents\LGMVIP Internship\78969efd3ca5c81769d31337b140005d.jpg")
```

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
img.shape
```

```python
# grayscale
gray= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
plt.axis("off")
plt.title("GrayScale Image")
```

```python
# Blur
blur= cv2.GaussianBlur(img,(5,5),cv2.BORDER_DEFAULT)
plt.imshow(blur)
plt.axis("off")
plt.title("GaussianBlur Image")
```

```python
# Edge cascade
canny= cv2.Canny(img,125,255)
plt.imshow(canny)
plt.axis("off")
plt.title("Canny Image")
```

```python
# Edge cascade
canny2= cv2.Canny(blur,0,255)
plt.imshow(canny2)
plt.axis("off")
plt.title("Cannyblur Image")
```

```python
# Dilating the image
dilated = cv2.dilate(canny, (7,7), iterations=3)
plt.imshow(dilated)
plt.axis("off")
plt.title("Dilate Image")
```

```python
# resize
resized= cv2.resize(img,(100,300),interpolation=cv2.INTER_LINEAR)
resize=plt.imshow(resized)
plt.axis("off")
plt.title("Resize Image")
```

```python
# Cropping
cropped = img[30:150, 130:150]
crop=plt.imshow(cropped)
plt.axis("off")
plt.title("crop Image")
```

```python
ret, thresh = cv2.threshold(gray, 125, 255, cv2.THRESH_BINARY)  
plt.imshow(thresh)
plt.axis("off")
plt.title("thresh Image")
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

```python
invert = cv2.bitwise_not(gray)
plt.imshow(invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

```python
img_smoothing = cv2.GaussianBlur(invert, (21, 21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.show()
```

```python
#blur= cv2.GaussianBlur(img_invert,(5,5),cv2.BORDER_DEFAULT)
#plt.imshow(blur)
```

```python
final = cv2.divide(gray, 255-img_smoothing, scale=255)
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Sketch Image")
plt.show()
```
